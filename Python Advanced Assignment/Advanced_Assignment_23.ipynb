```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for two sets of figures
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting the figures
plt.plot(x, y1, label='Figure 1', color='blue', marker='o')
plt.plot(x, y2, label='Figure 2', color='green', marker='s')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Comparison of Two Figures')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

#### Q3. What is a histogram, exactly? Name a numpy method for creating such a graph ?

**Ans:** A histogram is a graphical representation of the distribution of a dataset. It displays the frequencies of observations or occurrences of values within specified ranges, often referred to as bins. The x-axis represents the range of values, and the y-axis represents the frequency of occurrences within each bin.

 In a histogram, data is divided into intervals, and the height of each bar corresponds to the frequency (or relative frequency) of data points falling into that interval.To create a histogram in Python using NumPy, you can use the numpy.histogram method. This method computes the histogram of a set of data and returns the frequency counts and bin edges.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
data = np.random.randn(1000)

# Create a histogram
hist, bins = np.histogram(data, bins=20)

# Plot the histogram
plt.hist(data, bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

#### Q4. If necessary, how do you change the aspect ratios between the X and Y axes ?
**Ans:** In Matplotlib, you can change the aspect ratio between the X and Y axes by using the plt.axis or ax.set_aspect methods. These methods allow you to control the scaling of the axes. The aspect ratio is defined as the ratio of the height to the width of the axes.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Plot the data
plt.scatter(x, y)

# Set aspect ratio to be equal
plt.axis('equal')

# Display the plot
plt.show()
```

In this example, plt.axis('equal') sets the aspect ratio to be equal, ensuring that the scaling along both the X and Y axes is the same.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a subplot
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y)

# Set aspect ratio to be equal
ax.set_aspect('equal')

# Display the plot
plt.show()
```

In this example, ax.set_aspect('equal') achieves the same result, but it is applied to an individual subplot.

#### Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays ?
**Ans:**  In NumPy, there are three main types of array multiplication:

#### Dot Product:
- **Operation:** Sum of products of corresponding elements in two arrays.
- **NumPy Function:** `numpy.dot` or `@` operator.
- **Use Case:** Common in linear algebra, matrix multiplication, and vector operations.
- **Result Shape:** Scalar for 1-D arrays, matrix multiplication for 2-D arrays.

#### Outer Product:
- **Operation:** Computes a matrix by multiplying each element of one vector with each element of another.
- **NumPy Function:** `numpy.outer`.
- **Use Case:** Used in calculating covariance matrices, tensor products, and outer products in linear algebra.
- **Result Shape:** Cartesian product of input array shapes.

#### Regular Element-wise Multiplication:
- **Operation:** Standard element-wise multiplication of corresponding elements in two arrays.
- **NumPy Operator:** `*`.
- **Use Case:** Applied when multiplying corresponding elements of arrays element-wise.
- **Result Shape:** Same as input array shapes.


#### Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment ?
**Ans:** To calculate the monthly mortgage payment, you can use the numpy function numpy.pmt. This function is commonly used to calculate loan payments based on fixed interest rates, loan amount, and loan term.

```python
import numpy_financial as npf

# Parameters for the mortgage
loan_amount = 300000  # Loan amount in dollars
annual_interest_rate = 0.04  # Annual interest rate (4%)
loan_term_years = 30  # Loan term in years

# Monthly interest rate
monthly_interest_rate = annual_interest_rate / 12

# Total number of monthly payments
total_payments = loan_term_years * 12

# Calculate monthly mortgage payment using numpy_financial.pmt
monthly_payment = npf.pmt(monthly_interest_rate, total_payments, -loan_amount)

print(f"Your monthly mortgage payment: ${monthly_payment:.2f}")
```
